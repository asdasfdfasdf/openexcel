```python
import numpy as np
import pandas as pd
```

```python
#定义借呗放款df
borrowing_lending_df = pd.read_csv(r'F:\Downloads\借呗下载数据\20220811\借呗放款.csv')
#替换金额中的逗号为空格
borrowing_lending_df['放款金额(元)'] = borrowing_lending_df['放款金额(元)'].str.replace(',', '').astype(float)
borrowing_lending_df
```

```python
#定义借呗还款df
loan_repayment_df = pd.read_csv(r'F:\Downloads\借呗下载数据\20220811\借呗还款明细.csv')
loan_repayment_df['本次应还总金额（元）'] = loan_repayment_df['本次应还总金额（元）'].replace(',', '', regex=True).astype(float)
loan_repayment_df['本次实还总金额（元）'] = loan_repayment_df['本次实还总金额（元）'].replace(',', '', regex=True).astype(float)
loan_repayment_df['还款本金(元)'] = loan_repayment_df['还款本金(元)'].replace(',', '', regex=True).astype(float)
loan_repayment_df
```

```python
#定义借呗代偿df
compensatory_df = pd.read_csv(r'F:\Downloads\借呗下载数据\20220811\借呗代偿明细.csv')
compensatory_df['理赔本金 (元)'] = compensatory_df['理赔本金 (元)'].replace(',', '', regex=True).astype(float)
compensatory_df['正常本金 (元)'] = compensatory_df['正常本金 (元)'].replace(',', '', regex=True).astype(float)
compensatory_df
```

```python
#定义借呗追回df
retrieve_df = pd.read_csv(r'F:\Downloads\借呗下载数据\20220811\借呗追偿明细.csv')
retrieve_df['应还总金额(元)'] = retrieve_df['应还总金额(元)'].replace(',', '', regex=True).astype(float)
retrieve_df['实还总金额（元）'] = retrieve_df['实还总金额（元）'].replace(',', '', regex=True).astype(float)
retrieve_df
```

```python
# 定义借呗还款剔除了追偿的df
loan_repayment_retrieve_df = loan_repayment_df[~loan_repayment_df['还款流水号'].isin(retrieve_df['还款流水号'])]
loan_repayment_df
```

```python
# 获取loan_repayment_retrieve_df根据借据号分组的数据还款本金和
loan_repayment_retrieve_df_group = loan_repayment_retrieve_df.groupby('借据号')['还款本金(元)'].sum().reset_index()
loan_repayment_retrieve_df_group
```

```python
# 左连接代偿本金
compensatory_df_group = compensatory_df.groupby('借据号')['理赔本金 (元)'].sum().reset_index()
# 左连接
loan_repayment_retrieve_df_group_left = pd.merge(loan_repayment_retrieve_df_group, compensatory_df_group, on='借据号',
                                                 how='left')
loan_repayment_retrieve_df_group_left.replace(np.nan, 0, inplace=True)
loan_repayment_retrieve_df_group_left['还款和代偿总和'] = (loan_repayment_retrieve_df_group_left['还款本金(元)'] +
                                                    loan_repayment_retrieve_df_group_left['理赔本金 (元)'])
loan_repayment_retrieve_df_group_left
```

```python
# 将borrowing_lending_df和loan_repayment_retrieve_df_group_left进行左连接
borrowing_lending_df_left = pd.merge(borrowing_lending_df, loan_repayment_retrieve_df_group_left, on='借据号',
                                     how='left')
borrowing_lending_df_left.replace(np.nan, 0, inplace=True)
# 创建在保df
in_df = borrowing_lending_df_left[borrowing_lending_df_left['还款和代偿总和'] != borrowing_lending_df_left['放款金额(元)']]
# 在保不良df
bad_in_df = in_df[in_df['理赔本金 (元)'] != 0]
bad_in_df
```

```python
sum((borrowing_lending_df_left['放款金额(元)'] - borrowing_lending_df_left['还款和代偿总和']))
```

```python
# 计算剩余代偿
sum(bad_in_df['放款金额(元)'] - bad_in_df['还款和代偿总和'])
```

```python
# 筛选还款本金为0的数据df
in_df_zero = in_df[in_df['还款本金(元)'] == 0]
in_df_zero
```

```python
sum(in_df_zero['放款金额(元)'] - in_df_zero['还款和代偿总和'])
```
